```python
from scipy.signal import chirp, spectrogram
import numpy as np
import matplotlib.pyplot as plt
```

```python
duration = 10
sampleRate = 48000
f_start = 10
f_end = 24000

t = np.linspace(0, duration, sampleRate*duration + 1) 
w = chirp(t, f0=f_start, f1=f_end, t1=duration, method='linear')
w_inv = chirp(t, f0=f_start, f1=f_end, t1=duration, method='linear')[::-1]
```

```python
pulse = np.convolve(w, w_inv, mode='full') 
```

```python
t = np.linspace(0, duration*2, sampleRate*duration*2 + 1)
plt.plot(t, pulse)
plt.title("Linear Chirp, f(0)=6, f(10)=1") 
plt.xlabel('t (sec)') 
plt.show()
```

```python
with open('sweep.fb', "wb") as f:
    f.write(w)
    f.close()
```

```python
with open('sweep.txt', "w") as f:
    f.write(str(w))
    f.close()
```
